```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  #bu 3 satırda kullanacağımız kütüphaneler tanımlanmıştır.

from google.colab import files

uploaded = files.upload()

upload_file = 'Housing.csv'
df = pd.read_csv('Housing.csv')
```

```python
plt.figure(figsize=(25, 10)) #grafik boytunu ayarlar.
plt.scatter(df['area'], df['price'], color="purple", edgecolor="black", linewidth=0.5 )
plt.title('Ev Fiyatlarına Göre Evlerin Metrekareleri')
plt.xlabel('Metrekare', fontsize=14)
plt.ylabel('Fiyat', fontsize=14)
plt.show()
```

```python
ortalama_fiyat = df.groupby('bedrooms')['price'].mean() #dataframedeki verileri bedrooms a göre gruplar ve sadece price dikkate alınır. mean ise price sütunundaki değerlerin ortalamasını hesaplar.

plt.figure(figsize=(5, 5))

plt.bar(ortalama_fiyat.index, ortalama_fiyat.values, color='pink', width=0.3, alpha=0.7, edgecolor='black', linewidth=1)
plt.title('Oda Sayısına Göre Ortalama Fiyat', fontsize=18)
plt.xlabel('Oda Sayısı', fontsize=14)
plt.ylabel('Ortalama Fiyat', fontsize=14)
plt.show()
```

```python
ortalama_fiyat = df.groupby('stories')['price'].mean() #dataframedeki verileri stories e göre gruplar ve sadece price dikkate alınır. mean ise price sütunundaki değerlerin ortalamasını hesaplar.

plt.figure(figsize=(5, 5))

plt.bar(ortalama_fiyat.index, ortalama_fiyat.values, color='red', alpha=0.7, width=0.3, edgecolor='black', linewidth=1)
plt.title('Kat Sayısına Göre Ortalama Fiyat', fontsize=18)
plt.xlabel('Kat Sayısı', fontsize=14)
plt.ylabel('Ortalama Fiyat', fontsize=14)
plt.show()
```

```python
bathrooms_price = df.groupby('bathrooms')['price'].mean() #dataframedeki verileri bathrooms a göre gruplar ve sadece price dikkate alınır. mean ise price sütunundaki değerlerin ortalamasını hesaplar.

plt.figure(figsize=(5, 5))

plt.bar(bathrooms_price.index, bathrooms_price.values, color='cyan', alpha=0.7, width=0.3, edgecolor='black', linewidth=1)
plt.title('Lavabo Sayısına Göre Ortalama Fiyat', fontsize=18)
plt.xlabel('Lavabo Sayısı', fontsize=14)
plt.ylabel('Ortalama Fiyat', fontsize=14)
plt.show()
```

```python
furnishingstatus = df.groupby('furnishingstatus')['price'].mean() #dataframedeki verileri furnishingstatus a göre gruplar ve sadece price dikkate alınır. mean ise price sütunundaki değerlerin ortalamasını hesaplar.

plt.figure(figsize=(5, 5))

plt.bar(furnishingstatus.index, furnishingstatus.values, color='yellow', alpha=0.7, width=0.3, edgecolor='black', linewidth=1)
plt.title('Furnishingstatus Durumuna Göre Ortalama Fiyat', fontsize=18)
plt.xlabel('Furnishingstatus Durumu', fontsize=14)
plt.ylabel('Ortalama Fiyat', fontsize=14)
plt.show()
```

```python
import seaborn as sns
correlation_matrix = df.corr(numeric_only=True) #korelasyon matriksini hesaplamamızı sağlar.

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='viridis', square=True, cbar_kws={"shrink": .8}) #annot hücreler içine korelasyon değerinin yazılmasını sağlar. fmt sayıların formatını belirler. cmap renk paletini belirler. square şeklini ayarlar. renk çubuğunun boyutunu ayarlar.
plt.title('Korelasyon Matrisi', fontsize=16)
plt.show()
```
